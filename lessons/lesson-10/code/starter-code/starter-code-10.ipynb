### Visualizing models over variables

```python
%matplotlib inline
```

```python
import pandas as pd
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../../assets/dataset/flight_delays.csv')
df = df.loc[df.DEP_DEL15.notnull()].copy()
```

```python
#how do you get the data dictionary
#Day of Week = categorical
df.dtypes
```

```python
df.head()
```

```python
df = df[df.DEP_DEL15.notnull()]
df = df.join(pd.get_dummies(df['CARRIER'], prefix='carrier'))
df = df.join(pd.get_dummies(df['DAY_OF_WEEK'], prefix='dow'))
model = lm.LogisticRegression()
features = [i for i in df.columns if 'dow_' in i]

#features... want a quick way of dragging only my dummy columns for 'Day of Week'
#features = returning a list of columns names based on the criteria
```

```python
df.shape
```

```python
#here we are adding a feature
features += ['CRS_DEP_TIME']

model.fit(df[features[1:]], df['DEP_DEL15'])

#.T means "transpose". This is done to make it a column
df['probability'] = model.predict_proba(df[features[1:]]).T[1]
```

```python
#these are our predictors
#the baseline is removed here
df[features[1:]].head()

#numpy array that includes a list
#predict_proba will predict based on point estimates
model.predict_proba(df[features[1:]])
```

```python
ax = plt.subplot(111)
colors = ['blue', 'green', 'red', 'purple', 'orange', 'yellow']

#enumerate assigns numbers to columns
#features is a list
#df[features[e]] == 1 will get the dataframe for every day of the week
for e, c in enumerate(colors):
    df[df[features[e]] == 1].plot(x='CRS_DEP_TIME', y='probability', kind='scatter', color = c, ax=ax)

ax.set(title='Probability of Delay\n Based on Day of Week and Time of Day')



#blue is monday, green is tuesday, red is wednesday, etc
#this graph tells us that for a given departure time, we can expect which day has the higher probability fo being delayed
```
